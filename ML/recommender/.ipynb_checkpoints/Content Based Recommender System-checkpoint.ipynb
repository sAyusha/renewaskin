```python
top_concerns = {'face-moisturisers':'general care', 'mask-and-peel':'daily use', 'cleanser':'general care', 'sunscreen':'sun protection' }
entries = len(df2)
for i in range(entries):
    label = df2.iloc[i]['label']
    if pd.isnull(df2.iloc[i]['concern']):
        df2.iloc[i]['concern'] = top_concerns[label]
df2.info()

    
```

```python
df2['brand'] = df2['brand'].str.lower()
df2['name'] = df2['name'].str.lower()
df2['skin type'] = df2['skin type'].str.lower()
df2['concern'] = df2['concern'].str.replace(' and ', ',').str.replace(' or ', ',')
df2
```

```python
def concern_elements(comma_sep_concerns):
    words = comma_sep_concerns.split(',')
    for w in words:
        if w != '':
            concerns.add(w.rstrip())


            
# features
list(df2['skin type'].unique())
concerns = set([])
for i in range(entries):
    concern_elements(df2.iloc[i]['concern'])
    
```

```python
concerns
```

```python
concerns.remove('anti acne scarring')
concerns.add('scarring')
concerns.remove('blackheads removal')
concerns.remove('dull skin')
concerns.remove('dullness')
concerns.add('dull')
concerns.remove('oily skin')
concerns.remove('excess oil')
concerns.remove('oil control')
concerns.add('oil')
```

```python
features = list(df2['skin type'].unique()) + list(concerns)
features
```

```python
len(features)
```

```python
def search_concern(target, i):
    if target in df2.iloc[i]['concern']:
        return True
    return False

one_hot_encodings = np.zeros([entries, len(features)])

#skin types first
for i in range(entries):
    for j in range(6):
        skin_type = features[j]
        if skin_type == df2.iloc[i]['skin type']:
            one_hot_encodings[i][j] = 1

#other features
for i in range(entries):
    for j in range(6, len(features)):
        feature = features[j]
        if feature in df2.iloc[i]['concern']:
            one_hot_encodings[i][j] = 1
            
    
```

```python
def cosine_similarity(a, b):
    cos_sim = np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))
    return cos_sim


    
```

```python
df2.iloc[2]
```

```python
import heapq
heap = []

capacity = 6
feature_vector = one_hot_encodings[2]
df_temp = df2[df2['label'] == 'face-moisturisers']
for i in range(len(df_temp)):
    x = df_temp.iloc[i]['name']
    idx = df2.index[df2['name'] == x][0]
    cs = cosine_similarity(feature_vector, one_hot_encodings[idx])
    if capacity > 0:
        heapq.heappush(heap, [cs, idx])
        capacity -= 1
    else:
        if heap[0][0] < cs:
            heapq.heappop(heap)
            heapq.heappush(heap, [cs, idx])
print(f"INPUT CONCERNS: {df2.iloc[2]['concern'] }")
print("")
print("Recommended products (1/4 - moisturisers): ")
elements = heapq.nlargest(5, heap)
for index, element in enumerate(elements):
    score = element[0]
    product_data = df2.iloc[element[1]]
    print(f"{index+1}) Score : {score}")
    print(f"{product_data['name']} | {product_data.concern}")
    
```
